## Droping Of Irrelavent Attributes


```python
# Here we are dropping columns which have same values in all rows via pandas dataframe
nunique = dataset.apply(pd.Series.nunique)
cols_to_drop = nunique[nunique == 1].index
dataset1 = dataset.drop(cols_to_drop, axis=1)
dataset1.head()
```

```python
dataset1.shape
```

```python
# From the given Dataset the EmploeeID Does not Impact the dataset and hence we are dropping the column EmployeeID
dataset1.drop("EmployeeID",axis=1,inplace=True)
```

```python
dataset1.isna().sum()
```

```python
# Droping the null values permanently because of less number of null values.
dataset1.dropna(inplace = True)
dataset1.isna().sum()
```
